```python
#checking missing values
customer_data.isnull().sum()
```

Having missing values for review rating makes sense because not every customer is going to leave a review

For my purposes it makes sense to fill those missing values not with the median of the column as a whole, but with the median of the column grouped by category. This will help account for the possibility of rating varying a lot between categories

```python
customer_data['Review Rating']= customer_data.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
customer_data.isnull().sum()
```

```python
customer_data.columns = customer_data.columns.str.lower()
customer_data.columns = customer_data.columns.str.replace(' ', '_')
```

```python
customer_data.columns
```

The purchase amount column name still needs to be fixed

```python
customer_data = customer_data.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
customer_data.columns
```

```python
#creating an age group column
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
customer_data['age_group']= pd.qcut(customer_data['age'], q=4, labels = labels)
```

```python
customer_data[['age','age_group']].head(10)
```

```python
# creating a purchase_freequency in days column using the freequency of purchases col 

frequency_dict = {
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months' :90

}
customer_data['purchase_frequency_days'] = customer_data['frequency_of_purchases'].map(frequency_dict)
```

```python
customer_data[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
customer_data[['discount_applied','promo_code_used']].head(10)
```

Discount applied and promo code used seem like the same column. But a company may give out a discount without a the customer needing to use a code, so I need to see if the columns are redundant or not

```python
(customer_data['discount_applied'] == customer_data['promo_code_used']).all()
```

The columns have the exact same data so there is no need to keep both

```python
customer_data = customer_data.drop('promo_code_used', axis=1)
```

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine

# load  username/password from .env file
load_dotenv()

# connecting to PostgreSQL
username = os.getenv("POSTGRES_USERNAME")
password = os.getenv("POSTGRES_PASSWORD")
host = os.getenv("POSTGRES_HOST")
port = os.getenv("POSTGRES_PORT")
database = os.getenv("POSTGRES_DATABASE")

engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# loading customer_data into PostgreSQL
table_name = "customer"
customer_data.to_sql(table_name, engine, if_exists="replace", index=False)
```

```python
pd.read_sql(f"SELECT * FROM {table_name} LIMIT 5", engine)
```
